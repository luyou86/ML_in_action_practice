```python
from numpy import  *

def read_data_from_file(filename):
    f = open(filename)
    train_data = []
    label = []
    
    for line in f.readlines():
        tem_str = line.strip().split('\t')
        tem_list = []
        for i in range(len(tem_str) -1):
            tem_list.append(float(tem_str[i]))
        train_data.append(tem_list)
        label.append(float(tem_str[len(tem_str) -1]))
    
    return train_data, label

train_data, label = read_data_from_file('ex0.txt')
print train_data

            
```

```python
def regression1(train_data, label):
    train_data_matrix = mat(train_data)
    label_matrix = mat(label).T
    x_T_and_X = train_data_matrix.T * train_data_matrix
    
    if linalg.det(x_T_and_X) == 0:
        print 'can not compute reverse'
        return 
    parameters = x_T_and_X.I* train_data_matrix.T*label_matrix
    return parameters
parameters = regression1(train_data, label)
print parameters
```

```python
import matplotlib.pyplot as plt

def plot1(train_data, label):
    fig = plt.figure()
    axes = fig.add_subplot(111)
    train_data_matrix = mat(train_data)
    label_matrix = mat(label).T
    parameters = regression1(train_data, label)
    #important 
    train_data_copy = train_data_matrix.copy()
    train_data_copy.sort(0)
    predict_y = train_data_copy * parameters
    
    axes.scatter(train_data_matrix[:,1].flatten().A[0], label_matrix[:,0].flatten().A[0])
    axes.plot(train_data_copy, predict_y)
    
    plt.show()
    
plot1(train_data, label) 
    
```

```python
def LWLR_regression(test_point, train_data, label, k=1.0):
    train_data_matrix = mat(train_data)
    label_matrix = mat(label)
    row_num, col_num = train_data_matrix.shape
    lwlr_parameters = mat(eye((row_num)))
    
    for i in range(row_num):
        diff = test_point - train_data_matrix[i,:]
        lwlr_parameters[i,i] = exp(diff * diff.T / (-2*k*k))
        
    reverse_item = train_data_matrix.T * lwlr_parameters * train_data_matrix
    if linalg.det(reverse_item) == 0:
        print 'can not compute...'
        return 
    parameters = reverse_item.I * (train_data_matrix.T * (lwlr_parameters * label_matrix.T))
    return test_point * parameters

print LWLR_regression(train_data[1], train_data, label)
```

```python
def ridge_regression(train_data, label, super_parameter=1.0):
    train_data_matrix = mat(train_data)
    label_matrix = mat(label)
    tem = train_data_matrix.T * train_data_matrix + super_parameter * mat(eye(shape(train_data_matrix)[1]))
    
    if linalg.det(tem) == 0:
        print 'can not compute inverse....'
        return 
    parameters = tem.I * train_data_matrix.T * label_matrix
    return parameters

def test_ridge_regree(x, y):
    x_mat = mat(x)
    y_mat = mat(y).T  
    # by column
    y_mean = mean(y_mat, 0)
    y_mat = y_mat - y_mean
    
    x_mean = mean(x_mat, 0)
    x_var = var(x_mat, 0)
    x_mat = x_mat - x_mean / x_var
    
    weights = zeros((30, shape(x_mat)[1]))
    for i in range(30):
        weights[i,:] = (ridge_regression(x_mat, y_mat, exp(i-10))).T
    return weights

   
abalone_train_data ,abalone_label = read_data_from_file('abalone.txt')
#print len(abalone_label)
ridge_weighs = test_ridge_regree(abalone_train_data, abalone_label)

fig = plt.figure()
axes = fig.add_subplot(111)
axes.plot(ridge_weighs)
plt.show()
    
```

```python
def compute_error(x,y):
    return sum((x-y)**2)


def stage_wise_regression(train_data, label, epllision = 0.1, iteration_num = 30):
    train_data_mat = mat(train_data)
    label_mat = mat(label)
    row_num, col_num =  shape(train_data_mat)
    weights = zeros((iteration_num, col_num))
    tem_weights = zeros((col_num ,1))
    best_weights = tem_weights.copy()
    for it in range(iteration_num):
        print tem_weights.T
        min_error = inf
        for i in range(col_num):
            for choose in [-1,1]:
                weight_change = tem_weights.copy()
                weight_change[i] = epllision * choose
                predict_y =  train_data_mat * weight_change
                error = compute_error(label_mat.A, predict_y.A)
                
                if error < min_error:
                    min_error = error
                    best_weights = weight_change
        tem_weights = best_weights.copy()
        weights[it,:] = tem_weights.T
    return weights

abalone_weight = stage_wise_regression(abalone_train_data, abalone_label,0.01,200)                

print abalone_weight


            
```

```python
def read_data_from_file(filename):
    f = open(filename)
    data = []
    for line in f.readlines():
        tem_str = line.strip().split('\t')
        tem_list = [float(tem_str[0]), float(tem_str[1])]
        data.append(tem_list)
    
    return data
ex00_data = read_data_from_file('ex00.txt')

print len(ex00_data)
        
```

```python
class treeNode():
    def __init__(self, feature, value, left_node, right_node):
        self.feature = feature
        self.value = value
        self.left_node = left_node
        self.right_node = right_node
    

    
```

```python
from numpy import *

def produce_leaf(data_set):
    return mean(data_set[:, -1])
```

```python
def compute_error(data_set):
    return var(data_set[:, -1]) * shape(data_set)[0]
```

```python
def binary_split_dataset(data_set, feature, value):
    matrix1 = data_set[nonzero(data_set[:,feature] > value)[0],:]
    matrix2= data_set[nonzero(data_set[:,feature] <= value)[0],:]
    return matrix1, matrix2
```

```python
def choose_feature(data_set_mat, leaf_produce_method=produce_leaf, error_compute=compute_error, parameters=[1,4]):
    min_error_decrease = parameters[0]
    min_sample_num = parameters[1]
    #data_set_mat = mat(data_set)
    row_num, col_num = shape(data_set_mat)
    data_error = compute_error(data_set_mat)
    choose_feat_index = 0
    choose_value = 0
    minist_error = inf
    
    if len(set(data_set_mat[:,-1].T.tolist()[0])) == 1:
        return None, leaf_produce_method(data_set_mat)
    
    for col in range(col_num):
        for row in range(row_num):
            mat1, mat2 = binary_split_dataset(data_set_mat, col, data_set_mat[row, col])
            if shape(mat1)[0] < min_sample_num and shape(mat2)[0] < min_sample_num:
                continue
            temp_error = error_compute(mat1) + error_compute(mat2)
            print 'mat1: ', mat1.shape
            if temp_error < minist_error:
                choose_feat_index = col
                choose_value = data_set_mat[row, col]
                minist_error = temp_error
    mat1, mat2 = binary_split_dataset(data_set_mat, choose_feat_index,choose_value )       
    if (data_error - minist_error < min_error_decrease):
        return None, leaf_produce_method(data_set_mat)
    if shape(mat1)[0] < min_sample_num and shape(mat2)[0] < min_sample_num:
        return None, leaf_produce_method(data_set_mat)
    
    return choose_feat_index, choose_value

```

```python
def create_tree(data_set_mat,leaf_produce_mehtod=produce_leaf, error_compute=compute_error, parameters=[1,4]):
    feature, value = choose_feature(data_set_mat, leaf_produce_mehtod, error_compute,parameters)
    if feature == None:
        return value
    tree = {}
    tree["feature"] = feature
    tree["value"] = value
    mat1, mat2 = binary_split_dataset(data_set_mat, feature, value)
    tree["left_node"] = create_tree(mat1, leaf_produce_mehtod, error_compute, parameters)
    tree['right_node'] = create_tree(mat2, leaf_produce_mehtod, error_compute, parameters)
    
    return tree

```

```python
def is_tree(object):
    return type(object).__name__ == 'dict'

def get_mean(tree):
    if not is_tree(tree["left_node"])  and not is_tree(tree["right_node"]):
        return (tree['left_node'] + tree['right_node']) /2
    if is_tree(tree["left_node"]):
        return get_mean(tree['left_node'])
    if is_tree(tree["right_node"]):
        return get_mean(tree["left_node"])
```

```python
def prune(train_tree, test_data_mat):
    if not is_tree(train_tree["left_node"]) and not is_tree(train_tree["right_node"]):
        left_mat, right_mat = binary_split_dataset(test_data_mat, train_tree["feature"], train_tree["value"])
        error_without_merget = sum(power((left_mat[:,-1] - train_tree["left_node"]),2) ) + \
            sum(power((right_mat[:,-1] - train_tree["right_node"]), 2))
        merge_mean = (train_tree["left_node"] + train_tree["right_node"]) / 2
        error_with_merge = sum(power((test_data_mat[:,-1] - merge_mean), 2))
        if error_with_merge < error_without_merget:
            return merge_mean
        else :
            return train_tree
    if is_tree(train_tree["left_node"]):
        left_mat, right_mat = binary_split_dataset(test_data_mat, train_tree["feature"], train_tree["value"])
        return prune(train_tree["left_node"], left_mat)
    if is_tree(train_tree["right_node"]):
        left_mat, right_mat = binary_split_dataset(test_data_mat, train_tree["feature"], train_tree["value"])
        return prune(train_tree["right_node"], right_mat)
    if shape(test_data_mat)[0] == 0:
        return get_mean(train_tree)
```

```python
def lineal_method(data_set_mat):
    row_num, col_num = shape(data_set_mat)
    x = mat(ones((row_num, col_num)))
    y = mat(zeros((row_num, 1)))
    x[:,1:col_num] = data_set_mat[:,0:col_num-1]
    y = data_set_mat[:,col_num-1]
    x_T_and_x = x.T * x
    
    if linalg.det(x_T_and_x) == 0:
        raise NameError('the matirxi can not compute inverse!!!!')
       

    weights = x_T_and_x.I *x.T * y
    return weights, x, y

def model_leaf(data_set_mat):
    weights, x, y = lineal_method(data_set_mat)
    return weights

def model_error_compute(data_set_mat):
    weights , x, y = lineal_method(data_set_mat)
    print 'weights: ', weights
    predict_y = x * weights
    return sum(power(y - predict_y, 2))


    
    
```

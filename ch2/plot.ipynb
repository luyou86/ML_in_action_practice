```python
from matplotlib import pyplot as plt
import random
import traceback

def plot_scatter(X1,Y1, X2, Y2, X3, Y3):
    fig = plt.figure()  #
    axs = fig.add_subplot(111)  #11axes1
    
    #
    axs.scatter(X1, Y1, label="11111111",s=20, c='red')
    axs.scatter(X2, Y2, label="22222222", s=30, c='green')
    axs.scatter(X3, Y3, label='33333333', s=40, c='blue')
    
    #
    plt.title('scatter plot')
    
    #
    plt.xlabel("Miles per year")
    plt.ylabel("Percentage it cost")
    
    #legend
    axs.legend(loc=2)
    
    #
    plt.show()
    
   

feature1 =[]
label1 = []
for i in range(0,100):
    x = random.randint(1,100)
    y = random.randint(1, 100)
    feature1.append(x)
    label1.append(y)
    
feature2 = []
label2 = []
for i in range(0, 100):
    x = random.randint(1, 100)
    y = random.randint(1, 100)
    feature2.append(x)
    label2.append(y)
    
feature3 = []
label3 = []
for i in range(0, 100):
    x = random.randint(1,100)
    y = random.randint(1, 100)
    feature3.append(x)
    label3.append(y)
    
plot_scatter(feature1, label1, feature2, label2, feature3, label3)
```

```python
def plot_line(X1, Y1):
    fig = plt.figure()
    axes = fig.add_subplot(111)
    
    axes.plot(X1, Y1, label='line1', c='red')
    
    plt.title('Plot of X and Y')
    
    plt.xlabel("x axis")
    plt.ylabel("y axis")
    
    #
    plt.xlim(0,10)
    plt.ylim(0,30)
    
    axes.legend(loc = 2)
    plt.show()
    
X1 = [1,2,3,4,5,6]
Y1 = [2,4,6,8,10,15]
plot_line(X1, Y1)
```

```python
import numpy as np
from numpy.linalg import norm
def get_data_from_normal_distribution(sample_number, mean, variance):
    sample_data_list = np.random.normal(mean, variance, sample_number)
    
    return sample_data_list
sample_data = get_data_from_normal_distribution(2000, 3, 0.1)
fig = plt.figure()
axes = fig.add_subplot(111)
axes.hist(sample_data, label="$mean = 3,var = 0.1$", bins = np.linspace(2,4,200),color = 'blue')
plt.title("Normal distribution")
plt.xlabel("X labels")
plt.ylabel("Y labels")

plt.show()

#print sample_data
```
